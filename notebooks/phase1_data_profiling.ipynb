# Phase 1 - Data Profiling
Interactive profiling for `house_buying_dec29th_2025.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = '../data/raw/cresht2606_vietnam-real-estate-datasets-catalyst/house_buying_dec29th_2025.csv'
df = pd.read_csv(csv_path)
df.shape
```

```python
missing = df.isna().mean().sort_values(ascending=False)
missing.head(10)
```

```python
price = pd.to_numeric(df['price_million_vnd'], errors='coerce')
area = pd.to_numeric(df['area_m2'], errors='coerce')
price_per_m2 = price / area

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].hist(price.dropna(), bins=60)
axes[0].set_title('price_million_vnd')
axes[1].hist(area.dropna(), bins=60)
axes[1].set_title('area_m2')
axes[2].hist(price_per_m2.replace([float('inf'), float('-inf')], pd.NA).dropna(), bins=60)
axes[2].set_title('price_per_m2')
plt.tight_layout()
```
